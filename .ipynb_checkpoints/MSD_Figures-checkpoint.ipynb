Subset dataframes for plotting

```python
#specific to this data, can live here unless you figure out how to make it general
#could possibly make a general 'add condition subset' function that can do what this and the following function do
def Add_SoundTypeCol(df, condition = 'condition'):
    df['sound_type'] = (df.condition == 'Hcong') | (df.condition == 'HinCong')|(df.condition == 'Lcong') | (df.condition == 'LinCong')
    df['sound_type'] = df['sound_type'].astype(int)
    df['sound_type'] = df['sound_type'].replace({1: 'Animal', 0: 'White Noise'})
    return df

MSD = Add_SoundTypeCol(MSD)
MSDall = Add_SoundTypeCol(MSDall)
```

```python
#can probably still be cleaned up
def Add_SoundLocCol(df, condition = 'condition'):
    df['sound_loc1'] = (df['condition'] == 'Hcong') | (df['condition'] == 'HcongWN')
    df['sound_loc1'] = df['sound_loc1'].astype(int)
    df.sound_loc1 = df.sound_loc1.replace({1: 'Duck1', 0: 'Frog1'})
    df['sound_loc2'] = (df['condition'] == 'Lcong') | (df['condition'] == 'LcongWN')
    df['sound_loc2'] = df['sound_loc2'].astype(int)
    df.sound_loc2 = df.sound_loc2.replace({1: 'Frog2', 0: 'Duck2'})
    df['sound_loc'] = df['sound_loc1']
    df.loc[[4,5,6,7], 'sound_loc'] = df['sound_loc2']
    idx_sl2 = df.index[(df['condition']=='Lcong') |(df['condition']=='LcongWN')|(df['condition']=='LinCong')|(df['condition']=='LinCongWN')]
    df.loc[idx_sl2, 'sound_loc'] = df['sound_loc2']
    df = df.drop(columns=['sound_loc1', 'sound_loc2'])
    return df

MSD = Add_SoundLocCol(MSD)
MSDall = Add_SoundLocCol(MSDall)
```

```python
#should be made more general
def FirstFixProportions(countDF):
    cond_list = ['Hcong', 'HcongWN','HinCong', 'HinCongWN','Lcong', 'LcongWN','LinCong', 'LinCongWN']
    result = pd.DataFrame()
    for cond in cond_list:
        target = (countDF[('trial', cond)]['targetIA'])/((countDF[('trial',cond)]['distIA'])+
                                                          (countDF[('trial', cond)]['targetIA']))
        distractor = (countDF[('trial', cond)]['distIA'])/((countDF[('trial', cond)]['distIA'])+
                                                            (countDF[('trial', cond)]['targetIA']))
        PropDict = {'condition':cond,'target': target,'distractor':distractor}
        result = result.append(PropDict, ignore_index =True)
    return result

def FirstFixProportions(countDF, conditions, trialVar = 'trial', IA_1, IA_2 (or * type of thing here)):
    cond_list = ['Hcong', 'HcongWN','HinCong', 'HinCongWN','Lcong', 'LcongWN','LinCong', 'LinCongWN']
    result = pd.DataFrame()
    for cond in cond_list:
        target = (countDF[('trial', cond)]['targetIA'])/((countDF[('trial',cond)]['distIA'])+
                                                          (countDF[('trial', cond)]['targetIA']))
        distractor = (countDF[('trial', cond)]['distIA'])/((countDF[('trial', cond)]['distIA'])+
                                                            (countDF[('trial', cond)]['targetIA']))
        PropDict = {'condition':cond,'target': target,'distractor':distractor}
        result = result.append(PropDict, ignore_index =True)
    return result

FirstFix_DF = FirstFixProportions(MSD_firstfixDFCounts)
FirstFix_DF['sound_type'] = (FirstFix_DF['condition'] == 'Hcong') | (FirstFix_DF['condition'] == 'HinCong')|(FirstFix_DF['condition'] == 'Lcong') | (FirstFix_DF['condition'] == 'LinCong')
FirstFix_DF['sound_type'] = FirstFix_DF['sound_type'].astype(int)
FirstFix_DF.sound_type = FirstFix_DF.sound_type.replace({1: 'Animal', 0: 'White Noise'})
FirstFix_DF
```
